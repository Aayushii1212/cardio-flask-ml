<h1>Model train</h1>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
```

```python
x=df.drop(columns='cardio')
y=df['cardio']
x.head(),y.head()
```

```python
corr=df.corr()
```

```python
import seaborn as svs
svs.heatmap(corr)
```

```python
# x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
type(x_train)
```

```python
num_cols = ['ageyears','heightm','weight','ap_hi','ap_lo','bmi','pulse_pressure']
df.columns.tolist()
```

```python
# from sklearn.preprocessing import StandardScaler
# scaler=StandardScaler()
# x_train=scaler.fit_transform(x_train[num_cols])
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train[num_cols])

x_train.columns.tolist()
# x_train.shape
```

```python
x_train[num_cols] = scaler.transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

```python
model=LogisticRegression(max_iter=2000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
# y_prob=model.predict_proba(x_test)[:,1]
# y_prob 
```

```python
model.score(x_test,y_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```python
# roc_auc_score(y_test,y_prob)
```

```python
from sklearn.ensemble import RandomForestClassifier

# rf = RandomForestClassifier(
#     n_estimators=300,
#     max_depth=None,
#     random_state=42,
#     n_jobs=-1
# )
rf = RandomForestClassifier(
    n_estimators=600,
    max_depth=12,
    min_samples_split=20,
    min_samples_leaf=10,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

rf.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='auc',
    tree_method="hist"   # makes it fast
)

xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

xgb.score(x_test,y_test)
# from sklearn.metrics import accuracy_score, roc_auc_score
# print("XGB Accuracy:", accuracy_score(y_test, y_pred))
# print("XGB AUC:", roc_auc_score(y_test, xgb.predict_proba(x_test)[:,1]))
```
